## data preprocessing

```python
data1=data.copy()
```

```python
data1.head()
```

### Scale the data

```python
data.columns
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
data1[['age', 'gender', 'blood_type', 'medical_condition',
       'insurance_provider', 'billing_amount', 'admission_type', 'medication']]=scaler.fit_transform(data1[['age', 'gender', 'blood_type', 'medical_condition',
       'insurance_provider', 'billing_amount', 'admission_type', 'medication']])
```

```python
data1 # scaling the data
```

## Create X and Y

```python
X = data1.drop('test_results',axis=1)
y = data1.test_results
```

```python
X.head()
```

```python
y.head()
```

### Train and Test Split

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=56)
```

```python
x_train.shape
```

```python
y_test.shape
```

## SMOTE

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
smote=SMOTE()
x_smote,y_smote=smote.fit_resample(x_train,y_train)
print('Actual Value',Counter(y_train))
print('SMOTE Value',Counter(y_smote))
      
```

## Create the models

## Model -1 Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier()
dt.fit(x_smote,y_smote)
dt_predict = dt.predict(x_test)
```

```python
## Check Metrics Score
from sklearn.metrics import accuracy_score,classification_report
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_error
```

```python
accuracy_score(y_test,dt_predict)
```

```python
r2_score(y_test,dt_predict)
```

```python
print(classification_report(y_test,dt_predict))
```

## Model -2 Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
random = RandomForestClassifier()
random.fit(x_smote,y_smote)
rand_predict = random.predict(x_test)
```

```python
accuracy_score(y_test,rand_predict)
```

```python
r2_score(y_test,rand_predict)
```

```python
print(classification_report(y_test,rand_predict))
```
